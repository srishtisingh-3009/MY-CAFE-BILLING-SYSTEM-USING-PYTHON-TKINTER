```python
from tkinter import *
from tkinter import messagebox
import import_ipynb

root2=Tk()
root2.geometry("200x200+0+0")
root2.title("To Change the Price")
root2.iconbitmap("C:\\Users\\dell")

l=Label(root2,font=('helvetica',52,'bold'),text="Please Change the Prices of Items",fg="#0b034a",bd=10,bg="#e8de9e",anchor='w')
l.grid(row=0,columnspan=5,padx=20,pady=20)

fries1=StringVar()
pizza1=StringVar()
momos1=StringVar()
burger1=StringVar()
icecream1=StringVar()
softdrink1=StringVar()

def update():
    file=open('price_values.txt','r')
    line=file.readlines()
    fries_p=float(line[0])
    pizza_p=float(line[1])
    momos_p=float(line[2])
    burger_p=float(line[3])
    icecream_p=float(line[4])
    softdrink_p=float(line[5])
    file.close()
    
    file1=open('price_values.txt','w')
    try:
        cof1=float(fries1.get())
    except Exception as e:
        if fries1.get()!="":
            messagebox.showinfo('Error','Please enter the correct input!')
        fries1.set("")
        file1.write(str(fries_p)+"\n")
    else:
        file1.write(str(cof1)+"\n")
        
    try:
        cop1=float(pizza1.get())
    except Exception as e:
        if pizza1.get()!="":
            messagebox.showinfo('Error','Please enter the correct input!')
        pizza1.set("")
        file1.write(str(pizza_p)+"\n")
    else:
        file1.write(str(cop1)+"\n")
    
    try:
        com1=float(momos1.get())
    except Exception as e:
        if momos1.get()!="":
            messagebox.showinfo('Error','Please enter the correct input!')
        momos1.set("")
        file1.write(str(momos_p)+"\n")
    else:
        file1.write(str(com1)+"\n")
        
    try:
        cob1=float(burger1.get())
    except Exception as e:
        if burger1.get()!="":
            messagebox.showinfo('Error','Please enter the correct input!')
        burger1.set("")
        file1.write(str(burger_p)+"\n")
    else:
        file1.write(str(cob1)+"\n")
        
    try:
        coi1=float(icecream1.get())
    except Exception as e:
        if icecream1.get()!="":
            messagebox.showinfo('Error','Please enter the correct input!')
        icecream1.set("")
        file1.write(str(icecream_p)+"\n")
    else:
        file1.write(str(coi1)+"\n")
        
    try:
        cos1=float(softdrink1.get())
    except Exception as e:
        if softdrink1.get()!="":
            messagebox.showinfo('Error','Please enter the correct input!')
        softdrink1.set("")
        file1.write(str(softdrink_p)+"\n")
    else:
        file1.write(str(cos1))
    messagebox.showinfo('Update Box','Successfully Updated')
    file1.close()

def reset1():
    fries1.set("")
    pizza1.set("")
    burger1.set("")
    momos1.set("")
    icecream1.set("")
    softdrink1.set("")

def exit1():
    root2.destroy()
    

def printfn():
    file2=open('price_values.txt','r')
    liness=file2.readlines()
    fries_p=float(liness[0])
    pizza_p=float(liness[1])
    momos_p=float(liness[2])
    burger_p=float(liness[3])
    icecream_p=float(liness[4])
    softdrink_p=float(liness[5])
    print("===================")
    print("")
    print("fry: "+str(fries_p))
    print("pizza: "+str(pizza_p))
    print("momos: "+str(momos_p))
    print("burger:"+str(burger_p))
    print("icecream: "+str(icecream_p))
    print("softdrink: "+str(softdrink_p))
    print("")
    print("===================")
    file2.close()
    
def backfn():
    root2.destroy()
    import nextstep
    

f=Label(root2,font=('helvetica',16,'bold'),text="Fries:",bd=16,anchor='w')
f.grid(row=1,column=0,sticky=E)
ftxt1=Entry(root2,font=('helvetica',16,'bold'),textvariable=fries1,bd=10,insertwidth=4,bg="powder blue",justify="right")
ftxt1.grid(row=1,column=1,pady=10)

p=Label(root2,font=('helvetica',16,'bold'),text="Pizza:",bd=16,anchor='w')
p.grid(row=2,column=0,sticky=E)
ptxt1=Entry(root2,font=('helvetica',16,'bold'),textvariable=pizza1,bd=10,insertwidth=4,bg="powder blue",justify="right")
ptxt1.grid(row=2,column=1)

m=Label(root2,font=('helvetica',16,'bold'),text="Momos:",bd=16,anchor='w')
m.grid(row=3,column=0,sticky=E)
mtxt1=Entry(root2,font=('helvetica',16,'bold'),textvariable=momos1,bd=10,insertwidth=4,bg="powder blue",justify="right")
mtxt1.grid(row=3,column=1)

b=Label(root2,font=('helvetica',16,'bold'),text="Burger:",bd=16,anchor='w')
b.grid(row=4,column=0,sticky=E)
btxt1=Entry(root2,font=('helvetica',16,'bold'),textvariable=burger1,bd=10,insertwidth=4,bg="powder blue",justify="right")
btxt1.grid(row=4,column=1)

i=Label(root2,font=('helvetica',16,'bold'),text="Icecream:",bd=16,anchor='w')
i.grid(row=5,column=0,sticky=E)
itxt1=Entry(root2,font=('helvetica',16,'bold'),textvariable=icecream1,bd=10,insertwidth=4,bg="powder blue",justify="right")
itxt1.grid(row=5,column=1)

s=Label(root2,font=('helvetica',16,'bold'),text="Soft Drinks:",bd=16,anchor='w')
s.grid(row=6,column=0,sticky=E)
stxt1=Entry(root2,font=('helvetica',16,'bold'),textvariable=softdrink1,bd=10,insertwidth=4,bg="powder blue",justify="right")
stxt1.grid(row=6,column=1)

btn_update=Button(root2,padx=16,pady=8,fg="white",bd=10,font=('helvetica',16,'bold'),width=10,text="Update",bg="#6b130d",command=update)
btn_update.grid(row=7,column=0)

btn_reset1=Button(root2,padx=16,pady=8,fg="white",bd=10,font=('helvetica',16,'bold'),width=10,text="Reset",bg="#6b130d",command=reset1)
btn_reset1.grid(row=7,column=1)

btn_exit1=Button(root2,padx=18,fg="white",pady=8,bd=10,font=('helvetica',16,'bold'),width=10,text="Exit",bg="#6b130d",command=exit1)
btn_exit1.grid(row=7,column=2)

btn_print=Button(root2,padx=20,pady=8,fg="white",bd=10,font=('helvetica',16,'bold'),width=10,text="Print",bg="#6b130d",command=printfn)
btn_print.grid(row=7,column=3)

btn_back=Button(root2,padx=18,pady=8,fg="white",bd=10,font=('helvetica',16,'bold'),width=10,text="Back",bg="#6b130d",command=backfn)
btn_back.grid(row=7,column=4)


root2.mainloop()
```
